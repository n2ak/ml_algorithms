```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
import sklearn.datasets
import sklearn.metrics
import sklearn.model_selection
import sklearn.tree


seed = 0
X, y = sklearn.datasets.make_regression(n_features=1, noise=10, random_state=seed)
y += 10
a = np.linspace(X.min(), X.max(), 100)[:, None]
import sklearn.model_selection
from decision_tree import DecisionTree
class CustomDecisionTreeRegressor(DecisionTree):
    _criterions = ["squared_error"]

    def score(self,X,y):
        pred = self.predict(X)
        return sklearn.metrics.r2_score(y, pred)

model = sklearn.tree.DecisionTreeRegressor().fit(X,y)
custom = CustomDecisionTreeRegressor("squared_error").fit(X,y)
plt.plot(a, model.predict(a), color="g", label="Sklearn")
plt.plot(a, custom.predict(a), color="r", label="Our")
plt.scatter(X, y, label="data")
plt.legend()
```

```python
import sklearn.datasets
import sklearn.metrics
import sklearn.model_selection
X, y = sklearn.datasets.make_regression(n_features=5, noise=10, random_state=seed)
y += 10
sklearn.model_selection.cross_val_score(model, X,y),sklearn.model_selection.cross_val_score(custom, X,y)
```
